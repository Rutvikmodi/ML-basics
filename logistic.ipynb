```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
marks_df = pd.read_csv("C:/Users/Admin/Desktop/marks.csv")
```

```python
marks_df.head()
```

```python
marks_df.columns=['Subject 1','Subject 2','Pass or Fail']
marks_df.to_csv('marks.csv',index=False)
```

```python
marks_df.head()
```

```python
X = marks_df.iloc[:, :1]
```

```python
X
```

```python
 y = marks_df.iloc[:, -1]
```

```python
passed = marks_df.loc[y == 1]
```

```python
passed
```

```python
failed = marks_df.loc[y==0]
failed
```

```python
plt.scatter(passed.iloc[:, 0], passed.iloc[:, 1], s=10, label='Passed')
plt.scatter(failed.iloc[:, 0], failed.iloc[:, 1], s=10, label='Failed')
plt.legend()
plt.show()
```

```python
from sklearn.cross_validation import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
logreg.fit(X_train,y_train)
y_pred=logreg.predict(X_test)
```

```python
from sklearn import metrics
cnf_matrix = metrics.confusion_matrix(y_test, y_pred)
cnf_matrix
```

```python
import seaborn as sns
class_names=[0,1] # name  of classes
fig, ax = plt.subplots()
tick_marks = np.arange(len(class_names))
plt.xticks(tick_marks, class_names)
plt.yticks(tick_marks, class_names)
# create heatmap
sns.heatmap(pd.DataFrame(cnf_matrix), annot=True, cmap="YlGnBu" ,fmt='g')
ax.xaxis.set_label_position("top")
plt.tight_layout()
plt.title('Confusion matrix due to subject 1', y=1.1)
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
```
